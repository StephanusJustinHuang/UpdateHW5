```python
import numpy as np
import scipy.sparse as sp
from scipy.sparse import csr_matrix
from pyamg.gallery import poisson
from pyamg.relaxation.relaxation import jacobi, gauss_seidel
save = 0
```

```python
grid = (10,10)
A = poisson(grid).tocsr()
x_true = np.ones(A.shape[0],)
b = A.dot(x_true)
x = np.random.rand(A.shape[0],)

n_procs = 4
rows_per_proc = 25

A_on_procs = list()
A_off_procs = list()

for proc in range(n_procs):    
    on_ptr = list()
    off_ptr = list()
    on_cols = list()
    off_cols = list()
    on_data = list()
    off_data = list()
    on_ptr.append(0)
    off_ptr.append(0)
    
    start = proc*rows_per_proc
    end = (proc+1)*rows_per_proc
    for row in range(start):
        on_ptr.append(0)
        off_ptr.append(0)
    for row in range(start, end):
        for j in range(A.indptr[row], A.indptr[row+1]):
            col = A.indices[j]
            if (col >= start and col < end):
                on_cols.append(col)
                on_data.append(A.data[j])
            else:
                off_cols.append(col)
                off_data.append(A.data[j])
        on_ptr.append(len(on_cols))
        off_ptr.append(len(off_cols))
    for row in range(end, A.shape[0]):
        on_ptr.append(len(on_cols))
        off_ptr.append(len(off_cols))
        
    A_on_procs.append(csr_matrix((on_data, on_cols, on_ptr), shape=(100,100)))
    A_off_procs.append(csr_matrix((off_data, off_cols, off_ptr), shape=(100,100)))
```

```python
def hybrid_jacobi_gs(A_on_procs, A_off_procs, x, b):
    for proc in range(n_procs):
        gauss_seidel(A_on_procs[proc], x, b)
        jacobi(A_off_procs[proc], x, b)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.animation import FuncAnimation

def update_plot(frame_num, surf, x_true, b_procs, x):
    if frame_num > 0:
        hybrid_jacobi_gs(A_on_procs, A_off_procs, x, b_procs)
        
    error = np.abs(x_true - x)
    E = np.reshape(error, grid)
        
    surf[0].remove()
    surf[0] = ax.plot_surface(X, Y, E, cmap=cm.coolwarm)
    ax.set_zlim(0, 0.6)
    
    return surf
```

```python
from IPython.display import HTML
fig = plt.figure()
ax = plt.axes(projection='3d')
    
X = np.arange(0, grid[0])
Y = np.arange(0, grid[1])
X, Y = np.meshgrid(X, Y)

error = np.abs(x_true - x)
E = np.reshape(error, grid) 
surf = [ax.plot_surface(X, Y, E, cmap=cm.coolwarm)]
anim = FuncAnimation(fig, update_plot, 15, fargs=(surf, x_true, b, x), interval=1000)
```
